# Transform

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(
    {
        "groups":["A","B","C","A","B","C"],
        "var1" : [10,23,33,22,11,99],
        "var2" : [100,253,333,262,111,23],
    },
    columns = ["groups","var1","var2"]
)
df
```

```python
df["var1"]*2
```

```python
df
```

```python
df_a = df.iloc[:,1:3]
df_a
```

```python
df_a.transform(lambda x: x-x.mean())
```

```python
df_a.transform(lambda x: (x-x.mean()) / x.std())
```
