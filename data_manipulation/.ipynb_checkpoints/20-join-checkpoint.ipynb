# Birleştirme join işlemleri

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30,(10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

```python
df.head()
```

```python
df2 = df+99
df2
```

```python
pd.concat([df,df2]) # birleştirdik ancak index'lere bakarsak bi yanlışlık olduğunu görürüz
```
