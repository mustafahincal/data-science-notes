# Heatmap (Isı Haritası)

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df = df.pivot("month","year","passengers")
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df,annot = True,fmt = "d");
```

```python
sns.heatmap(df,annot = True,fmt = "d",linewidths=0.6);
```

```python
sns.heatmap(df,annot = True,fmt = "d",linewidths=0.6, cbar = False);
```
