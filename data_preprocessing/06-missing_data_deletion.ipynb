# Eksik Veri Silme Yöntemleri

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)
df
```

## İçinde herhangi bir değişken değerinde NaN olanları silelim

```python
df.dropna()
```

## Sadece tüm değişken değerleri NaN olanları silelim

```python
df.dropna(how ="all")
```

## Herhangi bir satırında NaN içeren sütunu silelim

```python
df.dropna(axis = 1) # default axis = 0'dır => satır bazlı düşünür. axis = 1'de değişkeni yani sütunu siliyoruz.
```

## Tüm değerleri NaN olan sütunu(değişkeni) silelim.

```python
df.dropna(axis = 1, how = "all")
```

```python
df["var_to_delete"] = np.nan
df
```

```python
df.dropna(axis = 1, how = "all", inplace = True)
```
