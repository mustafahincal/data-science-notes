# Eksik Veri Yapısının Görselleştirilmesi

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)
df
```

```python
!pip install missingno
```

```python
import missingno as msno
```

```python
msno.bar(df);
```

### yukarıdaki görselde, sağda değişkenlerdeki eleman sayılarını gösteriyor, yukarıda boş olmayan eleman sayılarını gösteriyor.

```python
msno.matrix(df);
```

### yukardaki görselde, boşluklar boş değerleri gösteriyor örneğin, ilk satıra bakarsak V3'ün boş olduğunu anlarız. Sağda ise 3'se eğer 3 değişkenin de boş olmadığını, 0'sa eğer 3 değişkeninde o satırın boş olduğunu anlıyoruz.

```python
import seaborn as sns
planets = sns.load_dataset("planets")
planets.head()
```

```python
planets.isnull()
```

```python
planets.isnull().sum()
```

```python
msno.matrix(planets);
```

### yukarıdan anlayabiliriz ki orbital_period ile mass arasında bir ilişki var.

```python
msno.heatmap(planets);
```

### yukarıdaki görselde değişkenlerin birbiriyleriyle olan ilişkilerinin güçlüklerini görüyoruz aslında, orbital_period ile mass arasında olduğu gibi distance ile mass arasında da bir ilişki var. Diğer görselde bunu pek saptayamamıştık.
