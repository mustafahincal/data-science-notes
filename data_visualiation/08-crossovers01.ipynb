```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
from pandas.api.types import CategoricalDtype
cut_categories = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_categories, ordered = True))
df.head()
```

```python
sns.catplot(x = "cut", y ="price", data=df);
```

```python
sns.barplot(x = "cut", y ="price", data=df);
```

```python
sns.barplot(x = "cut", y ="price", hue = "color", data=df); # siyah Ã§ubuklar standart sapma deÄŸerleri
```

```python
df.groupby(["cut","color"])["price"].mean() # deÄŸerleri incelediÄŸimiz zaman grafikle uyumlu olduÄŸunu gÃ¶rÃ¼yoruz
```
