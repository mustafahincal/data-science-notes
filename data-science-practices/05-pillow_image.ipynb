```python
import os
```

```python
folder_path = "./img"
paths = os.listdir(folder_path)
```

```python
paths
```

```python
for path in paths:
    if path.endswith(".png"):
        img = Image.open(os.path.join(folder_path,path))
        thumb = img.copy()
        thumb.thumbnail((500,300))
        thumb = thumb.transpose(method=Image.Transpose.FLIP_LEFT_RIGHT)
        thumb.save(f"./updated_img/{path}")
```
