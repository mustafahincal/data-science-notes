```python
import pandas as pd
```

```python
df = pd.read_csv("./data/titanic.csv")
df.head()
```

```python
df.drop(columns=["Cabin","PassengerId","SibSp","Parch","Ticket","Fare","Embarked"], inplace=True)
# df.drop(labels=["Cabin","PassengerId","SibSp","Parch","Ticket","Fare","Embarked"], axis = 1, inplace=True)  
df.head()
```

```python
cols = ["Name","Pclass","Age","Sex","Survived"]
df = df[cols]
df.head()
```

```python
df[df["Sex"] != "male"].index # female indexes
```

```python
df.drop(df[df["Sex"] != "male"].index)
```

```python
df["PClass"] = df["Pclass"]
df.drop(columns=["Pclass"], inplace=True)
df.head()
# Pclass -> PClass
```

```python
def add_st(x):
      return str(x) + "st"

df['PClass'] = df['PClass'].apply(add_st)
df.head()

# df['PClass'] = df['PClass'].apply(lambda x: str(x) + 'st')
```

```python
df.drop_duplicates() # it drops line that has same value in all columns
```

```python
df.drop_duplicates(subset=["Sex"])
```

```python
df.drop_duplicates(subset=["Sex"], keep="last")
```

```python
data_a = {
    "id" : ["1","2","3"],
    "name" : ["Mustafa","Patrick","Borini"],
    "surname" : ["Hincal","Kayne","Sanio"]
}
data_b = {
    "id" : ["4","5","6"],
    "name" : ["Robert","Jane","Jacob"],
    "surname" : ["Mor","Can","Lopez"]
}
```

```python
df_a = pd.DataFrame(data_a)
df_a
```

```python
df_b = pd.DataFrame(data_b)
df_b
```

```python
df_concat = pd.concat([df_a,df_b], axis = 0)
df_concat
```

```python
# df_concat.index = [1,2,3,4,5,6]
df_concat = pd.concat([df_a,df_b],ignore_index=True)
df_concat
```

```python
df_1 = pd.DataFrame({
                        "workers":["Jacob","Emily","Jason","Mark"],
                        "group":["Engineering","Engineering","Business Management","HR"]
                    })
df_1
```

```python
df_2 = pd.DataFrame({
                        "workers":["Jacob","Emily","Jason","Mark"],
                        "entry_date":[2010,2009,2014,2005]
                    })
df_2
```

```python
df_merge = pd.merge(df_1,df_2)
df_merge
```

```python
df_merge = pd.merge(df_1,df_2, on="workers")
df_merge
```

```python
df_merge = pd.merge(df_1,df_2, on="workers",how="outer")  # how="right","left" we can use sql join things here
df_merge
```
