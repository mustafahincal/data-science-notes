```python
import pandas as pd
```

```python
df = pd.read_csv("./data/titanic.csv")
df.head()
```

```python
df.drop(columns=["Cabin","PassengerId","SibSp","Parch","Ticket","Fare","Embarked"], inplace=True)
# df.drop(labels=["Cabin","PassengerId","SibSp","Parch","Ticket","Fare","Embarked"], axis = 1, inplace=True)  
df.head()
```

```python
cols = ["Name","Pclass","Age","Sex","Survived"]
df = df[cols]
df.head()
```

```python
df[df["Sex"] != "male"].index # female indexes
```

```python
df.drop(df[df["Sex"] != "male"].index)
```

```python
df["PClass"] = df["Pclass"]
df.drop(columns=["Pclass"], inplace=True)
df.head()
# Pclass -> PClass
```

```python
def add_st(x):
      return str(x) + "st"

df['PClass'] = df['PClass'].apply(add_st)
df.head()

# df['PClass'] = df['PClass'].apply(lambda x: str(x) + 'st')
```

```python
df.drop_duplicates() # it drops line that has same value in all columns
```

```python
df.drop_duplicates(subset=["Sex"])
```

```python
df.drop_duplicates(subset=["Sex"], keep="last")
```
