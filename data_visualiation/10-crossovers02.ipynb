# Çaprazlamalar

```python
import seaborn as sns
diamonds_data_set = sns.load_dataset("diamonds")
df = diamonds_data_set.copy()
df.head()
```

```python
sns.kdeplot(df.price,shade=True); # burda y ekseni çok önemli değil aslında sadece yoğunluğun nerelerde olduğunu anlamamız yeterli
```

```python
# peki bu mavi alanı oluşturan nedir?
```

```python
(sns
     .FacetGrid(df,
               hue="cut",
               height=5
                )
     .map(sns.kdeplot,"price",shade = True)
     .add_legend()
    );
```

```python
(sns
     .FacetGrid(df,
               hue="cut",
               height=5,
                xlim=(0,10000)
                )
     .map(sns.kdeplot,"price",shade = True)
     .add_legend()
    );
```

```python
sns.catplot(x = "cut", y = "price", hue = "color", data=df);
```

```python
sns.barplot(x = "cut", y ="price", hue = "color", data=df);
```

```python
sns.catplot(x = "cut", y = "price", hue = "color", kind="point", data=df);
```
