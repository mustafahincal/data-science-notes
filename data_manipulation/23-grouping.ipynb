# Gruplama (Grouping)

```python
import pandas as pd
df = pd.DataFrame({
                    "groups" : ["A","B","C","A","B","C"],
                    "data": [10,11,52,23,43,55]
            }, columns = ["groups","data"]
    )
df
```

```python
df.groupby("groups")
```

```python
df.groupby("groups").mean()
```

```python
df.groupby("groups").sum()
```

```python
import seaborn as sns
dfPlanets = sns.load_dataset("planets")
dfPlanets.head()
```

```python
dfPlanets.groupby("method").count()
```

```python
dfPlanets.groupby("method").sum() # dersek her değişkeni kendi arasında toplar ancak daha çok tek değişken için yazarız
```

```python
dfPlanets.groupby("method")["orbital_period"].sum()
```

```python
dfPlanets.groupby("method")["orbital_period"].describe()
```
