# One - Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
df.head()
```

```python
import pandas as pd
```

```python
df_one_hot = pd.get_dummies(df,columns = ["sex"],prefix = ["sex"])
```

```python
df_one_hot.head()
```

```python
# görüldüğü üzere sex_Male ve sex_Female olmak üzere iki column eklendi.
```

### Dummy Değişken Tuzağı = Oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa bu duruma dummy değişken tuzağı denir. sex_Male değişkeni sex_Female değişkeni üzerinden oluşturulabiliyor. Yani ikisi de aynı şeyi ifade ediyor.

### Kategorik değişkenin sınıf sayısından daha az sayıda dummy değişken olması lazım yani bu örnekte 1 tane değişken olmalı.

```python
pd.get_dummies(df,columns = ["day"],prefix = ["day"])
```

### One - Hot dönüşümünü sadece çok sınıflı kategorik değişkenlerde kullanmamız doğru olur çünkü sınıf arttıkça, oluşan yeni değişkenlerin birbiri üzerinden oluşma durumu azalır yani aynı şeyi ifade eden birden fazla değişken olmaz. Ancak sadece 2 sınıflı bir kategorik değişkeni One - Hot ile yaparsak aynı şeyi ifade eden iki değişkenimiz olur orada LabelEncoder kullanmamız daha doğrudur.
