# Değişken Dönüşümleri (Data Dönüşümleri)

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

## 0 - 1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df.head()
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
df.head()
```
